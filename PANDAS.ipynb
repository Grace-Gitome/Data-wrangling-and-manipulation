pd.merge()
most flexible
many-to-one and many-to-many joins are possible
side-by-side merge

```python
import pandas as pd
df1=pd.DataFrame({'ID':[1,2,3,5,9],
                  'col_1':[1,2,3,4,5],
                  'col_3':[6,7,8,9,10],
                  'col_3':[11,12,13,14,15],
                  'col_4':['apple','orange','banana','strawberry','rasberry']
                  })
df2=pd.DataFrame({'ID':[1,1,3,5],
                'col_A':[8,9,10,11],
                'col_B':[12,13,15,17],
                'col_4':['apple','orange','banana','kiwi']
                })
```

```python
df1
```

```python
df2
```

Concatination-for combining dataframes across rows and columns
Stacking a dataframe over the other

```python
pd.concat([df1,df2])
```

```python
inner=pd.merge(df1,df2) #This is same as specifying all columns that share a name btwn the 2 dataframes
inner
```

```python
#we can also specify to merge on a single column
pd.merge(df1,df2,on='col_4')
```

```python
#Merge on indices
pd.merge(df1,df2,suffixes=['_l','_r'],left_index=True,right_index=True)
```

Joins in pandas
types of joins
INNER-Joins all shared rows on the joining/key columns.You'll lose rows that dont have a match
in the other DF key column

OUTER-Joins all the rows from both the dataframe.No data will be lost.

LEFT-Joins all the rows from the left dataframe.Any rows from the right dataframe that do not have match in the key column of the left DF are discarded.

RIGHT-The opposite of left join.Joins all the rows from the right dataframe.Any rows from the left dataframe that do not have match in the key column of the right DF are discarded.

```python
#outer join
pd.merge(df1,df2,on='col_4',how='outer',suffixes=['_l','_r'])
```

```python
#left join
pd.merge(df1,df2,on='col_4',how='left',suffixes=['_l','_r'])
```

```python
#right join
pd.merge(df1,df2,on='col_4',how='right',suffixes=['_l','_r'])
```

Reshaping and pivot tables in pandas
Pivot tables allows you to transform or reshape dataframe

```python
df=pd.read_csv('Weather.csv')
df
```

```python
#pivot function
#reshaping
df.pivot(index='date',columns='city')
```

```python
#We can specify on the values we want to show
df.pivot(index='date',columns='city',values='humidity')
```

PIVOT TABLES
Pivot table is used to summarize and aggregate data inside dataframe 

```python
df=pd.read_csv('Weather_2.csv')
df
```

```python
df.pivot_table(index='city',columns='date',aggfunc='sum')
```
